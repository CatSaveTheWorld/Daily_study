**쉬운 설명**

아래에 알고리즘1, 2를 사용하여 나타낸 결과표와 크로스 엔트로피를 계산한 식이 있다고 하자.

클래스 설명
*   democrat : 민주당
*   republican : 공화당
*   other : 그외

<br>


**[Algorithm 1]**

|computed|targets|correct?|
|-|-|-|
|0.3 0.3 0.4|0 0 1 (democrat)|yes|
|0.3 0.4 0.3|0 1 0 (republican)|yes|
|0.1 0.2 0.7|1 0 0 (other)|no|


```
accuracy_score = 0.67
Cross Entropy = -(ln(0.4)) + ln(0.4) + ln(0.1)) / 3
              = 1.38
```
```python
# 파이썬 코드
import numpy as np

-(np.log(0.4) + np.log(0.4) + np.log(0.1)) / 3    # 1.3783888522474517
```

<br>

**[Algorithm 2]**

|computed|targets|correct?|
|-|-|-|
|0.1 0.2 0.7|0 0 1 (democrat)|yes|
|0.1 0.7 0.2|0 1 0 (republican)|yes|
|0.3 0.4 0.3|1 0 0 (other)|no|

```
accuracy_score = 0.67
Cross Entropy = -(ln(0.7)) + ln(0.7) + ln(0.3)) / 3
              = 0.64
```
```python
# 파이썬 코드
import numpy as np

-(np.log(0.7) + np.log(0.7) + np.log(0.3)) / 3    # 0.6391075640678003
```

<br>

여기서 알고리즘 1과 2를 비교하면 2의 성능이 더 높다고 할 수 있다.

왜냐하면 알고리즘 1의 경우 민주당[democrat] / 공화당[republican]의 분류 확률이 각각 0.4 / 0.4로 비교적 애매한 값으로 결과를 내고, 그 외 정당[other]의 경우 확률이 0.1로 아주 확실하게 틀린 결과를 냈다.

그러나, 알고리즘 2의 경우 민주당[democrat] / 공화당[republican]의 분류 확률이 각각 0.7 / 0.7로 확실한 값으로 결과를 내고, 그 외 정당[other]의 경우 확률이 0.3으로 애매한 값으로 틀린 답을 내놓았다.

이 설명을 **정량적인 수치로 나타낸 것이 바로 크로스 엔트로피[Cross Entropy]**이다.
각 표 아래의 식은 크로스 엔트로피를 계산한 식이다.

실제로, 분류의 결과는 알고리즘 1, 2가 동일하게 Accuracy Score가 0.67이지만, 크로스 엔트로피의 값을 보면 더 정확하고 확실한 확률로 분류를 한 알고리즘 2의 값이 더 낮게 나오는 것을 볼 수 있다.

따라서, **크로스 엔트로피는 값이 낮을수록 좋다.**

````python
# [Algorithm 1]
import numpy as np

-(np.log(0.4) + np.log(0.4) + np.log(0.1)) / 3    # 1.3783888522474517
````

````python
# [Algorithm 2]
import numpy as np

-(np.log(0.7) + np.log(0.7) + np.log(0.3)) / 3    # 0.6391075640678003
````
